```python
import matplotlib.pyplot as plt
import numpy as np
```

**SIMPLE BAR GRAPH**

```python
x=np.array(["A","B","C","D"])
y=np.array([20,40,10,50])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("BAR GRAPH")
plt.bar(x,y)
plt.show()
```

**BAR GRAPH WITH COLOURS**


```python
x=np.array(["A","B","C","D"])
y=np.array([20,40,10,50])
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.title("BAR GRAPH")
plt.bar(x,y,color="orange")
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([20,40,10,50])
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.title("BAR GRAPH")
plt.bar(x,y,color="purple")
plt.show()
```

**HORIZONTAL GRAPHS**

```python
#horizontal bargraph
x=np.array(["A","B","C","D"])
y=np.array([100,50,230,123])

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar Graph")
plt.barh(x,y,height=0.1,color="yellow")
plt.show()
```

```python
#bar graph with bar width
x=np.array(["A","B","C","D"])
y=np.array([100,50,230,123])

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar Graph")
plt.bar(x,y, width=0.4,color="pink")
plt.show()
```
